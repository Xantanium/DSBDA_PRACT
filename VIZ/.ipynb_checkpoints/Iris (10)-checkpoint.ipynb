```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')

# eda
print(df.head())
print(df.info())
```

```python
# list feature types:
print("\nFeature Types: ")
print(df.dtypes)

# check unique categoricals
print("\nUnique values in species: ")
print(df['species'].unique())
```

```python
# Histograms for each feature

df.hist(figsize=(10, 8), bins=20, edgecolor='black', color='skyblue')
plt.suptitle("Feature Distros - Iris Dataset")
plt.tight_layout()
plt.show()
```

```python
# Box plots for each feature
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=df, y=column, x='species')
    plt.title(f"Boxplot for {column}")

plt.tight_layout()
plt.show()
```
