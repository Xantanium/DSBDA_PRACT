```python
import pandas as pd
import numpy as np
```

```python
# ID deleted dataset
df = pd.read_csv("../DATASETS/IRIS.csv")
```

```python
print(df.head())
print()

# Group by the categorical column 'Species'
grouped = df.groupby(['species'])

# For each group, calculate summary statistics manually
for name, group in grouped:
    print(f"\nStatistics for {name}:")
    print("Mean:\n", group.mean(numeric_only=True))
    print("Median:\n", group.median(numeric_only=True))
    print("Minimum:\n", group.min(numeric_only=True))
    print("Maximum:\n", group.max(numeric_only=True))
    print("Standard Deviation:\n", group.std(numeric_only=True))
```
